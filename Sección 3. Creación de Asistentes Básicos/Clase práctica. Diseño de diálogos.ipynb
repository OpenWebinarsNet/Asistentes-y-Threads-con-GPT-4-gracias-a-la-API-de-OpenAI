### Claridad y concisión

El asistente debe emitir respuestas sencillas y fácilmente comprensibles para el usuario:

```python
instrucciones_precisas = """
Eres un profesor de Python para principiantes. Los usuarios te preguntarán sobre
temas relacionados con programación y tú debes ayudarlos. Para ello les explicarás
brevemente la teoría detrás de los conceptos implicados, la sintaxis general para
resolver la duda y por último darás un ejemplo de código comentando en cada línea
de código lo que estás haciendo. Solo responderás a dudas sobre Python, si te preguntan
sobre cualquier otro tema contestarás: *Lo siento, solo puedo resolverte dudas
sobre Python*. Tus respuestas deben ser cortas y evitar un lenguaje demasiado formal.
"""
```

```python
asistente_actualizado = client.beta.assistants.update(
  "asst_SM7wzJRb1DwuaDQd4wGl0EzN",
  name="Profesor de Python",
  instructions = instrucciones_precisas)
```

### Contexto

Necesitamos que el asistente sea capaz de recordar interacciones anteriores para que la comunicación sea natural y fluída. Esta parte se gestiona automáticamente por los threads. Para crear un hilo usamos el método `threads.create`. Veamos un ejemplo:

```python
hilo_vacio = client.beta.threads.create()
print(hilo_vacio)
```

Añadimos un mensaje al hilo:

```python
mensaje_hilo = client.beta.threads.messages.create(
  hilo_vacio.id,
  role="user",
  content="¿Cómo se multiplican números enteros?",
)
print(mensaje_hilo)
```

```python
ejecucion = client.beta.threads.runs.create(
  thread_id=hilo_vacio.id,
  assistant_id=asistente_actualizado.id
)

print(ejecucion)
```

Observando la conversación con la función de la clase anterior:

```python
def recupera_conversacion_de_hilo(hilo):
  conversacion = {}
  mensajes_hilo = client.beta.threads.messages.list(hilo)
  longitud_hilo = len(mensajes_hilo.data)
  for indice in range(0, longitud_hilo):
    informacion_mensaje = mensajes_hilo.data[longitud_hilo-1-indice]
    conversacion[indice] = (informacion_mensaje.role, informacion_mensaje.content[0].text.value)

  return conversacion

```

```python
conversacion_hilo = recupera_conversacion_de_hilo(hilo_vacio.id)
conversacion_hilo
```

Si preguntamos ahora en el hilo que tenemos en el simulador "¿y los decimales?" el asistente no será capaz de responder correctamente porque no tiene el contexto.

Sin embargo si añado el mensaje a este hilo:

```python
mensaje_hilo_2 = client.beta.threads.messages.create(
  hilo_vacio.id,
  role="user",
  content="¿Y los decimales?",
)
print(mensaje_hilo_2)
```

```python
ejecucion_2 = client.beta.threads.runs.create(
  thread_id=hilo_vacio.id,
  assistant_id=asistente_actualizado.id
)

print(ejecucion_2)
```

```python
conversacion_hilo_2 = recupera_conversacion_de_hilo(hilo_vacio.id)
conversacion_hilo_2
```

Podemos observar cómo se ha conservado el contexto de manera correcta permitiendo:

* Una interacción muy natural.
* Que interacciones en distintos hilos no interfieran entre ellas.

### Naturalidad

Para hacer que el diálogo sea más natural podemos añadir información a la petición, como por ejemplo, que siempre nos responda en el mismo idioma en el que preguntamos.

```python
naturalidad = "Responde siempre al usuario en el mismo idioma en el que te pregunta"
instrucciones_naturales = instrucciones_precisas + naturalidad
```

Actualizamos de nuevo el asistente:

```python
asistente_actualizado = client.beta.assistants.update(
  "asst_SM7wzJRb1DwuaDQd4wGl0EzN",
  instructions = instrucciones_naturales)
```

La empatía resulta clave para aportar naturalidad a las interacciones con asistentes virtuales. Una opción es, por ejemplo, sugerir que el asistente nos dé un mensaje de ánimo:
